# Monitoring Drug Shortages to Improve Medicine Supply Chain

Drug shortages can disrupt patient care and pose serious health risks.  
This project analyzes FDA drug shortage data to identify which drugs are most affected, the reasons for shortages, and which therapeutic areas are at highest risk.  

## What We Do

- Examine the **Top 10 drugs** with the most shortage records  
- Analyze the **Top 10 therapeutic categories** affected by shortages  
- Identify the **Top 10 reasons** for drug shortages  
- Explore trends in drug shortages over time. 

## Methodology

1. **Data Collection**: Pull data from the FDA Drug Shortage API  
2. **Data Cleaning**: Handle missing values, explode list-type columns, and convert dates  
3. **Exploratory Data Analysis (EDA)**:  
   - Visualize top drugs, therapeutic categories, and shortage reasons  
   - Plot trends over time to identify patterns  
4. **Insights**: Summarize key findings that can help healthcare providers, pharmacists, and policymakers anticipate shortages and make informed decisions.


```python
# Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from tqdm import tqdm
import requests
import os

sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Get the FDA API URL 

URL = "https://api.fda.gov/drug/shortages.json"

def fetch_shortages(limit=50, skip=0):
    """
    We fetch drug shortage data from the FDA API.
    
    Parameters:
    - limit: number of records per request (max 100)
    - skip: number of records to skip (for pagination)
    
    Returns:
    - JSON dictionary of results
    """
    url = f"{URL}?limit={limit}&skip={skip}"
    response = requests.get(url)

    if response.status_code != 200:
        print("Error:", response.status_code)
        return None

    return response.json()
```

```python
all_results = []  # empty list to store all records
limit = 100       # max records per API call

# Loop to get multiple pages of data
for skip in tqdm(range(0, 2000, limit)):  # adjust 2000 if you want more
    data = fetch_shortages(limit=limit, skip=skip)
    
    if data is None or "results" not in data:
        break
    
    all_results.extend(data["results"])
    
    # Stop if this is the last page
    if len(data["results"]) < limit:
        break

print(f"Total records fetched: {len(all_results)}"
```

```python
# We get the csv

df = pd.json_normalize(all_results)  # flatten JSON
csv_path = "../data/processed/fda_shortages.csv"
df.to_csv(csv_path, index=False)

print(f"CSV is saved to {csv_path}")
df.head()  # see first few rows
```

# Data Visualization

# To see drugs that appear most frequently

```python
plt.figure(figsize=(12,6))
top_drugs = df['generic_name'].value_counts().head(15)
sns.barplot(x=top_drugs.values, y=top_drugs.index, palette="viridis")
plt.title("Top 10 Drugs with Most Shortages")
plt.xlabel("Number of Shortages")
plt.ylabel("Drug Name")
plt.show()
```

# We are looking to visualize how drug shortages change over a time period

```python
# Convert date column to datetime
df['update_date'] = pd.to_datetime(df['update_date'], errors='coerce')

# Count shortages per month
shortages_over_time = df.groupby(df['update_date'].dt.to_period('M')).size()

# Plot
plt.figure(figsize=(12,6))
shortages_over_time.plot(kind='line', marker='o')
plt.title("Drug Shortages Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Shortages")
plt.show()
```

```python
# We visualize shortage reason now

plt.figure(figsize=(12,6))

# Count top 10 reasons
top_reasons = df['shortage_reason'].value_counts().head(10)

# Create horizontal bar plot
sns.barplot(x=top_reasons.values, y=top_reasons.index, palette="magma")

plt.title("Top 10 Shortage Reasons")
plt.xlabel("Number of Records")
plt.ylabel("Reason")
plt.show()
```

```python
# Now we check the number of drugs that are available

plt.figure(figsize=(15,12))

# Count the availability status
availability_counts = df['availability'].value_counts()

# Create bar plot
sns.barplot(x=availability_counts.index, y=availability_counts.values, palette="coolwarm")

plt.title("Availability Status of Drugs")
plt.xlabel("Availability")
plt.ylabel("Number of Drugs")
plt.show()
```

```python
# Explode the therapeutic_category column
df_exploded = df.explode('therapeutic_category')

# Now do value_counts
top_categories = df_exploded['therapeutic_category'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette="cividis")
plt.title("Top 10 Therapeutic Categories with Most Shortages")
plt.xlabel("Number of Shortages")
plt.ylabel("Therapeutic Category")
plt.show()

'''
Since some of the values in our therapeutic_category were lists Seaborn could not hash it.
So ,we exploded therapeutic_category into different rows
```

# We use explode function to handle list-type columns to avoid any error to generate plots.

```python
# Explode shortage_reason column 
df_exploded_reason = df.explode('shortage_reason')

# Count top 10 reasons
top_reasons = df_exploded_reason['shortage_reason'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=top_reasons.values, y=top_reasons.index, palette="magma")
plt.title("Top 10 Shortage Reasons")
plt.xlabel("Number of Records")
plt.ylabel("Reason")
plt.show()
```

```python
# similar with generic_name

# Explode generic_name just in case
df_exploded_drug = df.explode('generic_name')

# Count top 10 drugs
top_drugs = df_exploded_drug['generic_name'].value_counts().head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=top_drugs.values, y=top_drugs.index, palette="viridis")
plt.title("Top 10 Drugs with Most Shortages")
plt.xlabel("Number of Shortages")
plt.ylabel("Drug Name")
plt.show()
```

### Top 10 Drugs with Most Shortages

This bar plot shows the drugs that are most frequently reported as being in shortage. 

- The drug **Lisdexamfetamine Dimesylate Capsule** appears most often, indicating a persistent supply issue.  
- This insight can help pharmacists and healthcare providers prioritize monitoring and alternative sourcing for these critical medications.


### Top 10 Shortage Reasons 

This horizontal bar chart shows the most common reasons why drugs are in shortage.  

- The most frequent reasons include **Other**, and **Demand Increase for the Drug** **Shortage of Active Ingredient**  
- Identifying the common causes of shortages can help policymakers and pharmaceutical companies address systemic problems.  
- Since the top reason is **Other**, this means the specific reason was not specified so this could be improved with better data collection.

### Top 10 Therapeutic Categories with Most Shortages


- The most affected categories are **Pediatric**, and **Anesthesia**  
- This insight helps healthcare providers focus monitoring and resource allocation*for high-risk therapeutic areas.
